```python
import matplotlib.pyplot as plt
import requests
import numpy as np
from datetime import datetime
import time as t
import json
```

```python
lat = []
long = []
stamp = []

for num in range(1,101):
    print(num)
    r = requests.get('http://api.open-notify.org/iss-now.json')
    loc = r.json()
    if loc["message"] == "success":
        latitude = loc['iss_position']['latitude']
        longitude = loc['iss_position']['longitude']
        time = datetime.fromtimestamp(loc["timestamp"])
        lat.append(latitude)
        long.append(longitude)
        stamp.append(time)
    t.sleep(10)
plt.plot(lat,long,"b.-")
plt.title("ISS Location")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.show()
    
```
